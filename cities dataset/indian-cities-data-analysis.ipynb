```python
# Load necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data_path = '/kaggle/input/4000-indian-cities-names-states-lat-lon/Indian Cities Geo Data.csv'
cities_df = pd.read_csv(data_path)

# Display basic information about the dataset
print("Dataset Information:")
print(cities_df.info())

# Display descriptive statistics for numerical columns
print("\nDescriptive Statistics:")
print(cities_df.describe())
```

```python
# Check for missing values
print("\nMissing Values:")
print(cities_df.isnull().sum())
```

```python
# Explore unique values in 'State' column
print("\nUnique States:")
print(cities_df['State'].unique())

# Explore unique values in 'City' column
print("\nUnique Cities:")
print(cities_df['Location'].nunique())
```

```python
# Explore the distribution of cities across states
plt.figure(figsize=(12, 6))
sns.countplot(y='State', data=cities_df, order=cities_df['State'].value_counts().index)
plt.title('Number of Cities per State')
plt.xlabel('Number of Cities')
plt.ylabel('State')
plt.show()
```

```python
# Calculate the average latitude and longitude for each state
avg_coordinates = cities_df.groupby('State')[['Latitude', 'Longitude']].mean().reset_index()
print("\nAverage Coordinates per State:")
print(avg_coordinates.head())

```

```python
# Visualize the distribution of cities on a map
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Longitude', y='Latitude', data=cities_df)
plt.title('Distribution of Cities in India')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
